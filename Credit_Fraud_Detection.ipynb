```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fraud.columns = (fraud.columns.str.strip()
                              .str.replace(r'\s+', ' ', regex=True)
                              .str.lower()
                              .str.replace(' ', '_'))
```

```python
drop_cols = [
'first','last','street','city','dob'
]

fraud = fraud.drop(columns=drop_cols, errors='ignore')
```

```python
fraud['trans_date_trans_time'] = pd.to_datetime(fraud['trans_date_trans_time'], errors='coerce')
fraud['trans_hour'] = fraud['trans_date_trans_time'].dt.hour
fraud['trans_weekday'] = fraud['trans_date_trans_time'].dt.weekday
fraud['is_weekend'] = fraud['trans_weekday'].isin([5, 6]).astype(int)
```

```python
fraud['amt_log'] = np.log1p(fraud['amt'])                               
q75 = fraud['amt'].quantile(0.75)
fraud['high_amount'] = (fraud['amt'] > q75).astype(int)                  
```

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # km
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    return 2 * R * np.arcsin(np.sqrt(a))

fraud['distance_km'] = haversine(fraud['lat'], fraud['long'], fraud['merch_lat'], fraud['merch_long'])
```

```python
q75_dist = fraud['distance_km'].quantile(0.75)
fraud['far_transaction'] = (fraud['distance_km'] > q75_dist).astype(int)
```

```python
fraud['amt_log'] = np.log1p(fraud['amt'])
```

```python
fraud['trans_date'] = fraud['trans_date_trans_time'].dt.date
fraud['daily_trans_count'] = fraud.groupby(['cc_num', 'trans_date'])['amt'].transform('count')
fraud['daily_avg_amt'] = fraud.groupby(['cc_num', 'trans_date'])['amt'].transform('mean')
```

```python
cust_avg_amt = fraud.groupby('cc_num')['amt'].transform('mean')
fraud['amt_vs_cust_avg'] = fraud['amt'] - cust_avg_amt
```

```python
fraud['weekend_far'] = ((fraud['is_weekend'] == 1) & (fraud['far_transaction'] == 1)).astype(int)
```
